```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement_data.csv')
df = df.dropna()
x_values = df['CTC']
y_values = df['students']
df
```

```python
plt.scatter(x_values, y_values)
plt.xlabel('Number of Students')
plt.ylabel('CTC in LPA')
plt.show()
```

```python
x_values_array = x_values.to_numpy()
y_values_array = y_values.to_numpy()
x_values_array.reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(x_values_array, y_values_array, test_size=0.1, random_state=42)
X_train
```

```python
X_train = X_train.reshape(-1, 1)
X_train
X_test = X_test.reshape(-1,1)
model = LinearRegression()
model.fit(X_train, y_train)
```

```python

y_pred = model.predict(X_train)
y_pred
```

```python
plt.scatter(x_values_array, y_values_array, color='blue', label='Actual Data')
plt.plot(X_train, y_pred, color='red', linewidth=2, label='Regression Line')
plt.title('Linear Regression Fit')
plt.xlabel('Number of Students')
plt.ylabel('CTC in LPA')
plt.legend()
plt.show()
```
